```python
import pandas as pd

# Load dataset
df = pd.read_csv("../../data/raw/Hospital.csv")
df_copy = df.copy()

```

```python
print(df_copy.head())
print(df_copy.info())
print(df_copy.describe(include="all"))

# Missing values
print("\nMissing Values:")
print(df_copy.isnull().sum())

# Duplicate rows
print("\nDuplicate Rows:", df_copy.duplicated().sum())

# Data types
print("\nData Types:")
print(df_copy.dtypes)
```

```python
# Check current format
print(df_copy["Admission_Date"].head())
print(df_copy["Discharge_Date"].head())

# Convert to datetime
df_copy["Admission_Date"] = pd.to_datetime(df_copy["Admission_Date"], errors="coerce")
df_copy["Discharge_Date"] = pd.to_datetime(df_copy["Discharge_Date"], errors="coerce")

# Verify conversion
print(df_copy[["Admission_Date", "Discharge_Date"]].dtypes)

# Count invalid dates
print("Invalid Admission Dates:", df_copy["Admission_Date"].isna().sum())
print("Invalid Discharge Dates:", df_copy["Discharge_Date"].isna().sum())
```

```python
df_copy.to_csv("../../data/processed/Hospital_Processed_Date_Formats.csv")
```

```python
# Handle missing Discharge_Date values
df_copy["Treatment_Status"] = "Discharged"
df_copy.loc[df_copy["Discharge_Date"].isna(), "Treatment_Status"] = "Ongoing Treatment"

# Keep missing values as NULL for SQL import
df_copy["Discharge_Date"] = df_copy["Discharge_Date"].where(
    df_copy["Discharge_Date"].notna(), None
)
```

```python
# Standardize Medical_Condition
df_copy["Medical_Condition"] = (
    df_copy["Medical_Condition"]
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

# Standardize Department
df_copy["Department"] = (
    df_copy["Department"]
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)
```

```python
df_copy.to_csv("../../data/processed/Hospital_Clean_SQL_Import.csv", index=False)

print("Cleaning completed successfully!")
```

```python
print("\nFinal DataFrame Info:")
print(df_copy.info())   
```
